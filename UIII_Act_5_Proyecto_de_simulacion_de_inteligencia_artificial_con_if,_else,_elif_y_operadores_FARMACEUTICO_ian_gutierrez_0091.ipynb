```python
# ==========================================
# SISTEMA FARMACEUTICO CON "INTELIGENCIA ARTIFICIAL"
# Uso de if, elif, else y operadores
# Sin funciones, sin ciclos y sin POO
# ==========================================

print("===================================")
print(" FARMACIA INTELIGENTE ")
print("===================================")

nombre = input("Ingrese el nombre del paciente: ")
edad = int(input("Ingrese la edad: "))
temperatura = float(input("Ingrese la temperatura corporal: "))
tos = input("¿Tiene tos? (si/no): ")
dolor = input("¿Tiene dolor de cabeza? (si/no): ")
alergia = input("¿Es alergico a la penicilina? (si/no): ")
presion = input("¿Tiene presion alta? (si/no): ")

print("\nANALIZANDO SINTOMAS...\n")

# ------------------------------------------
# ANALISIS PRINCIPAL
# ------------------------------------------

if temperatura >= 39 and tos == "si" and dolor == "si":
    print("Posible infeccion fuerte.")
    print("Medicamento recomendado: Paracetamol + Jarabe")

    if alergia == "si":
        print("No administrar penicilina por alergia.")
    else:
        print("Puede usar antibiotico con supervision medica.")

elif temperatura >= 38 and tos == "si":
    print("Posible gripe.")
    print("Medicamento recomendado: Antigripal")

elif temperatura < 38 and dolor == "si":
    print("Posible migraña o cansancio.")
    print("Medicamento recomendado: Ibuprofeno")

elif temperatura >= 37 and temperatura < 38:
    print("Temperatura ligeramente elevada.")
    print("Se recomienda reposo e hidratacion.")

else:
    print("No se detectan sintomas graves.")
    print("Paciente estable.")

# ------------------------------------------
# ANALISIS EXTRA
# ------------------------------------------

if edad >= 60:
    print("Paciente de la tercera edad.")
    print("Se recomienda revision constante.")

elif edad < 12:
    print("Paciente menor de edad.")
    print("Usar dosis infantiles.")

if presion == "si":
    print("Advertencia: evitar medicamentos muy fuertes.")

# ------------------------------------------
# RESULTADO FINAL
# ------------------------------------------

print("\n===================================")
print(" REPORTE FINAL ")
print("===================================")

print("Paciente:", nombre)
print("Edad:", edad)
print("Temperatura:", temperatura)

print("programa realizado por ian gutierrez NL=26")
print("\nGracias por utilizar FARMACIA IA")
```
